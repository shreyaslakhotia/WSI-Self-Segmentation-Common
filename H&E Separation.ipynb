```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load Hematoxylin and Eosin images
hematoxylin_image = cv2.imread(r"C:\Users\lakho\Downloads\H1.jpg", cv2.IMREAD_GRAYSCALE)
eosin_image = cv2.imread(r"C:\Users\lakho\Downloads\E1.jpg", cv2.IMREAD_GRAYSCALE)

# Function to plot histogram
def plot_histogram(image, title, color):
    plt.hist(image.ravel(), bins=256, range=(0, 256), color=color, alpha=0.7)
    plt.title(f'Histogram - {title}')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

# Plot the images and histograms
plt.figure(figsize=(12, 8))

# Hematoxylin image
plt.subplot(2, 2, 1)
plt.imshow(hematoxylin_image, cmap='gray')
plt.title('Hematoxylin Image')
plt.axis('off')

# Hematoxylin histogram
plt.subplot(2, 2, 2)
plot_histogram(hematoxylin_image, 'Hematoxylin', 'blue')

# Eosin image
plt.subplot(2, 2, 3)
plt.imshow(eosin_image, cmap='gray')
plt.title('Eosin Image')
plt.axis('off')

# Eosin histogram
plt.subplot(2, 2, 4)
plot_histogram(eosin_image, 'Eosin', 'red')

plt.tight_layout()
plt.show()

```
